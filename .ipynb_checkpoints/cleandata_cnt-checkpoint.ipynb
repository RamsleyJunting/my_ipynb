### 拼接json（两个version）

```python
import os
def json_concat():
    base_path = "/world/data-c26/liliang/cleaned_data"
    last_version_path = os.path.join(base_path, "cleaned_v10_list.json")
    this_version_path = os.path.join(base_path, "cleaned_v11_list.json")
    out_json_path = os.path.join(base_path, "cleaned_v11_list_all.json")

    file_paths_tobe_concat = [last_version_path, this_version_path]
    total_rec = []

    for fp in file_paths_tobe_concat:
        rec = open(fp, "r").readlines()
        total_rec += rec
        print("%s: %s records"%(fp.split("/")[-1], len(rec)))

    with open(out_json_path, "w") as out_f:
        for line in total_rec:
            out_f.write(line)

    print("finish concating, total: %s records"%(len(total_rec)))

json_concat()
```

## Draw label2imgs distribution

```python
import json
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

def get_dataset_infos(json_path):
    record_dict = {}
    open_source_dataset = []
    if "labels2imgs" in json_path:
        f = open(json_path, "r")
        obj = json.load(f)
        for label, imgs in obj.items():
            record_dict[label] = len(imgs)
    with open(json_path, "r") as json_file:
        for idx, line in enumerate(json_file):
            # make record_dict to store all the needed information
            try:
                path, json_str = line.split("\t")
                obj = json.loads(json_str)
                label = obj["label"]
            except:
                continue
            # exclude all open source dataset
            if "/world/data-c7/reid_by_dataset" in path:
                open_source_dataset.append(line)
                continue
            if label in record_dict:
                record_dict[label] += 1
            else:
                record_dict[label] = 1

    total_ids = len(record_dict)
    return total_ids, record_dict, open_source_dataset

def get_scope(record_dict):
    label_lst = []
    scope = ["1", "2", "3", "4", "5-50", ">50"]
    scope_cnt = [0] * 6
    for label, cnt in record_dict.items():
        if cnt < 5:
            scope_cnt[cnt-1] += 1
        elif cnt > 50:
            scope_cnt[-1] += 1
        else:
            scope_cnt[-2] += 1
    print("distribution: %s"%dict(zip(scope, scope_cnt)))
    return scope, scope_cnt

def distribution(dataset_name, scope, scope_cnt, total_ids):
    plt.figure(figsize=(8,4))
    num_bins = 12
#     n, bins, patches = plt.hist(x, num_bins, normed=False, facecolor = 'blue', alpha = 0.3)
    plt.bar(scope, scope_cnt, width=0.5, color="green", align="center")
    plt.xticks(range(len(scope)), scope, rotation=45)
    plt.xlabel('label2imgs distribution')
    plt.ylabel('counts')
    plt.title(dataset_name + "   total_ids:%s"%total_ids)
    plt.show()
    
path = ["/world/data-c26/liliang/cleaned_data/cleaned_v6v7_list.json",
        "/world/data-c26/liliang/cleaned_data/cleaned_v9/labels2imgs.json",
#         "/world/data-c26/liliang/cleaned_data/v8_source_list.json",
#         "/world/data-c26/liliang/cleaned_data/cleaned_v8_list.json",
#           "/world/data-c26/liliang/cleaned_data/cleaned_v8_list.json",
          "/world/data-c26/liliang/cleaned_data/cleaned_v9_list.json"]
# path = ["/world/data-c7/person-reid-data/test_data/anta_chongqing.json",
#        "/world/data-c7/person-reid-data/test_data/anta_chongqing_cleaned.json"]
for p in path:
    dataset_name = p.split("/")[-1].split(".")[0]
    total_ids, record_dict, _ = get_dataset_infos(p)
    scope, scope_cnt = get_scope(record_dict)
    distribution(dataset_name, scope, scope_cnt, total_ids)
```
